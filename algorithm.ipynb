```python
import numpy as np
unique = []
train_file = open('pos_train.txt','r')
text = train_file.read()
words = text.split()
for word in words:
    if word not in unique:
        unique.append(word)
train_file.close()
train_file2 = open('pos_train.txt','r')
print(unique)
print(len(unique))
```

**Transition Probablity Matrix**

```python
dictionary_pos = {}   #dictionary for storing all POS and assigning unique serial for each
for i in range(len(unique)):
    dictionary_pos[i] = unique[i]
print(dictionary_pos)
transition_matrix = [[0]*13 for i in range(13)]  #transition matrix size is 13 as number of tags is 13
temp = 0
while temp!=1:
    x=train_file2.readline()
    if len(x)>0:
        wordlist = x.split(' ')
        for i in range(len(wordlist)-1):
            for j in range((len(dictionary_pos))):
                if(wordlist[i]==dictionary_pos.get(j)):
                    x_transition_matrix = j              #coordinate for x axis of transition matrix
                elif(wordlist[i+1]==dictionary_pos.get(j)):
                    y_transition_matrix = j              #coordinate for y axis of transition matrix
            transition_matrix[x_transition_matrix][y_transition_matrix]+=1  #incrementing for occurence of each pos for the set sequence
    else:
        temp = 1
transition_matrix_array = np.array(transition_matrix) 
print(transition_matrix_array)
print(transition_matrix_array.sum(axis =1)[:,None])
transition_matrix_array = transition_matrix_array/transition_matrix_array.sum(axis=1)[:,None] #this step is for calculating probablities for each transition
print(transition_matrix_array)
train_file2.close()
```

**Viterbi Algorithm**

```python
test_pos_file2 = open('pos_test.text','r')
test_token_file2 = open('tokens_test.txt','r')
len_found_sentences = 0
temp = 0
correct_prediction = 0
word_not_found = 0
while(temp!=1):  #while there is a sentence
    test_pos2 = test_pos_file2.readline()
    test_token2 = test_token_file2.readline()
    if len(test_pos2)>0:
        words_test_token2 = test_token2.split()
        words_test_pos2 = test_pos2.split()
        words_test_token2.remove('<s>') #logic of this algo is based on the sentence without start tag
        temp_list_current = []
        temp_list_current_index = []
        temp_list_previous = ['<s>']
        temp_list_previous_index = []
        predicted_pos_list_words = []
        biggest_prob= -1
        initial_prob = 1
        predicted_pos_list = [0]  #populating it initially by start tag
        r=0
        for comp in words_test_token2:
            if comp in unique_words_tokens:
                r+=1
        if(r==len(words_test_token2)):  #if any of the words is not in the train set, it will leave execution otherwise will continue
            len_found_sentences+=1   #tag for tracking number of sentences where words found in training data
            temp_list_current = []
            temp_list_current_index = [] #list for tracking current states
            temp_list_previous = ['<s>'] #list for tracking previous states
            temp_list_previous_index = []
            predicted_pos_list_words = []
            biggest_prob= -1
            initial_prob = 1
            predicted_pos_list = [0]
            for i in range(len(words_test_token2)):
                col = reversed_dictionary_token.get(words_test_token2[i])
                temp_arr = emission_matrix_array[:,col]
                for j in range(len(temp_arr)):
                    if(temp_arr[j]>0):
                        temp_list_current.append(dictionary_pos[j])
                for m in temp_list_previous:
                    temp_list_previous_index.append(reversed_dictionary_pos.get(m)) #getting serial number for each pos (previous)
                for n in temp_list_current:
                    temp_list_current_index.append(reversed_dictionary_pos.get(n)) #getting serial number for each pos (current)
                for p in temp_list_previous_index:
                    biggest_prob = -1
                    for q in temp_list_current_index:
                        temp_prob = transition_matrix_array[p][q]*temp_arr[q]*initial_prob #multiplying all probablities
                        if(temp_prob>biggest_prob): #viterbi core concept of selecting only the largest probablity
                            biggest_prob = temp_prob
                            x_p = p
                            y_q = q
                initial_prob = biggest_prob
                predicted_pos_list.append(y_q)                
                temp_list_previous = [dictionary_pos[y_q]]          
            
            for i in predicted_pos_list:
                predicted_pos_list_words.append(dictionary_pos[i])   
            if(words_test_pos2==predicted_pos_list_words):
                correct_prediction+=1    #if predictions are correct
        elif(r!= len(words_test_token2)): #if any of the words in the sentence is not present in training data
            word_not_found+=1
    else:
        temp = 1        
        
        
    
```
